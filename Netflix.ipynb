<h3>4. Cual es el TOP 10 por país en cuanto a cantidad de Producciones?</h3>

```python
dfNET['country'].value_counts()
```

```python
country = dfNET['country'].value_counts().head(10)

sns.barplot(x=country, y=country.index)
plt.xticks(rotation = 75)
plt.title('TOP 10 por país')
```

<p>Segun el set de datos analizado, se puede observar que el TOP 10 por país está compuesto por: Estados Unidos, India, Reino Unido, Japón, Corea del Sur,Canadá, España,Francia, México y Egipto.
<br><br>
    Siendo Estado Unidos el numero uno con un 41% del total de las producciones
</p>
