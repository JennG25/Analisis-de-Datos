```python
y=rawData["1 payment next month"]
y
```

```python
xClass=[]
xClass.append(("RFR",RandomForestRegressor()))
xClass.append(("LR",LinearRegression()))
xClass.append(("SVR",SVR()))
```

```python
resultado=[]
nombre=[]
for name, model in xClass:
    result=cross_val_score(model,x,y,cv=3,scoring="r2")
    nombre.append(name)
    resultado.append(result)
for i in range(len(nombre)):
    print(nombre[i], resultado[i].mean())
```
